```python
import pandas as pd
df=pd.read_csv('/content/Cutlets.csv')
df
```

```python
df['Unit A'].mean()
```

```python
df['Unit B'].mean()
```

```python

from scipy import stats

# Set the significance level
alpha = 0.05

# Extract the data from the two units
unit_a = df['Unit A'].values
unit_b = df['Unit B'].values

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(unit_a, unit_b)

# Make a decision based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis: There is significant difference in the mean cutlet weight between the two units.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the mean cutlet weight between the two units.")
```

```python
import pandas as pd
df=pd.read_csv('LabTAT.csv')
df
```

```python

import pandas as pd
from scipy import stats

# Set the significance level
alpha = 0.05

# Extract the data from the four laboratories
lab1 = df['Laboratory 1'].values
lab2 = df['Laboratory 2'].values
lab3 = df['Laboratory 3'].values
lab4 = df['Laboratory 4'].values

# Perform the one-way ANOVA test
f_statistic, p_value = stats.f_oneway(lab1, lab2, lab3, lab4)

# Make a decision based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the mean TAT among the four laboratories.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the mean TAT among the four laboratories.")
```

```python

import pandas as pd
import scipy.stats as stats

# Load the data
df = pd.read_csv('/content/Costomer+OrderForm.csv')

# Extract the data from the four centers
philippines = df[df['Phillippines'] == 1]
indonesia = df[df['Indonesia'] == 1]
malta = df[df['Malta'] == 1]
india = df[df['India'] == 1]

# Perform the chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency([[len(philippines), len(indonesia), len(malta), len(india)]])

# Set the significance level
alpha = 0.05

# Make a decision based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the percentage of defectiveness among the four centers.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the percentage of defectiveness among the four centers.")
```
